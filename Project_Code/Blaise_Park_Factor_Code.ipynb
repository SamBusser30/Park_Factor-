```python
# event_runs_ct
#totalruns_vs_altitude = HomeRunData['altitude'].value_counts()
#totalruns_vs_altitude1 = totalruns_vs_altitude.sort_index(axis=0, ascending=True, kind='quicksort').plot(kind='bar', title="Home Runs vs. Altitude")
#totalruns_vs_altitude1.set_ylabel("Number of Home Runs")
#totalruns_vs_altitude1.set_xlabel("Altitude")

total_runs = [[0 for i in range(2)] for j in range(30)] 

for i in range(0,30):
    total_runs[i][0] = altitudes['altitude'][i]
    

for i in range(0,30):
    
    for j in range(0,2111526):
    
        if events_with_altitudes['altitude'][j] == total_runs[i][0]:
        
            total_runs[i][1] += events_with_altitudes['event_runs_ct'][j]
    
    
total_runs
    
```
