```python
import pandas as pd 
import numpy as np
import matplotlib.pylab as plt
```

```python
#Upload the two files to a dataframe retrosheet has the baseball data while altitude has each stadiums altitude
Retrosheet = pd.read_csv('C://Users//knguy//Desktop//School//Spring 2018//Data Mining//Retrosheet.csv')
altitudes = pd.read_excel("C://Users//knguy//Desktop//School//Spring 2018//Data Mining//MLB_Stadium_Altitudes.xlsx")
```

```python
#Print out the first couple of lines
Retrosheet.head()
```

```python
#drop unnecessary attributes and set to a new dataframe
cleanRetro = Retrosheet.drop(['time_since_1900', 'wOBA_pts', 'bat_hand_cd', 'base1_run_id', 'base2_run_id', 'base3_run_id','pitch_seq_tx','game_new_fl','game_end_fl','po1_fld_cd', 'po2_fld_cd','pit_start_fl', 'pa_ball_ct', 'pa_strike_ct', 'bat_fate_id', 'run1_fate_id', 'run2_fate_id', 'run3_fate_id', 'battedball_cd', 'inn_new_fl', 'inn_end_fl', 'pit_hand_cd'], axis=1)

```

```python
#print out the new datafram that is now cleaned to a certain extent. More cleaning may occur
cleanRetro
```

```python
#now drop any unnecessary events we don't need 
#stolen bases (4), caught stealing (6), pickoff (8), 
print(len(Retrosheet))

cleanRetro = cleanRetro[cleanRetro.event_cd != (4 | 6 | 8)]

print(len(cleanRetro))
#Compare the lengths of the two dataframes to make sure it removed the events we didn't want
```

```python
#Merges the dataframe that is cleaned with the altitude dataframe so now we have the table with altitudes and we merge based on the home team id
events_with_altitudes = pd.merge(cleanRetro, altitudes, on='home_team_id')
```

```python
events_with_altitudes
```

```python
#Creating a dataframe that only has hit events 
# singles(20), doubles(21), triples(22), home runs(23)
HitsData1 = events_with_altitudes[events_with_altitudes['event_cd'].isin([20,21,22,23])]
```

```python
#Removing unnecessary colums since all I need is the home team name and the event
HitsData2 = HitsData1[['home_team_id', 'event_cd','altitude']]
HitsData2
```

```python
#Bar graph for hits vs stadium
hits_vs_stadium = HitsData2['home_team_id'].value_counts()
hits_vs_stadium1 = hits_vs_stadium.plot(kind='bar', title="Stadium vs. Hits")
hits_vs_stadium1.set_ylabel("Number of Hits")
hits_vs_stadium1.set_xlabel("Stadium")
```

```python
#graph for hits vs altitude
Hits_vs_altitude = HitsData2['altitude'].value_counts()
Hits_vs_altitude1 = Hits_vs_altitude.sort_index(axis=0, ascending=True, kind='quicksort').plot(kind='bar', title="Total Hits vs. Altitude")
Hits_vs_altitude1.set_ylabel("Number of Hits")
Hits_vs_altitude1.set_xlabel("Altitude")
```

```python
#Creating an At Bats datafram that consists of all the events that are included in an at bat such as stirkeouts. 
AtBatsData =  events_with_altitudes[events_with_altitudes['event_cd'].isin([2,3,14,15,16,18,19,20,21,22,23])]
```

```python
AtBatsData1 = AtBatsData[['home_team_id','event_cd','altitude']]
AtBatsData1
```

```python
#Doing a count based on home team id and setting that to a dataframe that counts total hits
HitsData3 = HitsData2['home_team_id'].value_counts().to_frame('Total Hits')
HitsData3.index.name = 'Stadium'
HitsData3.reset_index(inplace = True)
```

```python
#Doing a count for at bats
AtBatsData2 = AtBatsData1['home_team_id'].value_counts().to_frame('At Bats')
AtBatsData2.index.name = 'Stadium'
AtBatsData2.reset_index(inplace = True)
```

```python
HitsData3
```

```python
Merge = {'FLO':'MIA'}
HitsData3['Stadium'].replace(Merge,inplace=True)
HitsData3 = HitsData3.groupby('Stadium').sum().reset_index()
HitsData3.sort_values('Total Hits',ascending = False)
print(HitsData3)
```

```python
HitsData3.sort_values('Total Hits',ascending = False)
```

```python
AtBatsData2
```

```python
Merge = {'FLO':'MIA'}
AtBatsData2['Stadium'].replace(Merge,inplace=True)
AtBatsData2 = AtBatsData2.groupby('Stadium').sum().reset_index()
AtBatsData2.sort_values('At Bats',ascending = False)
print(AtBatsData2)
```

```python

AtBatsData2 = AtBatsData2.reset_index(drop=True)
AtBatsData2.sort_values('At Bats',ascending = False)
AtBatsData2
```

```python
#Merging at bats and total hits into a dataframe to calculate at bats. 
BattingAverage = pd.merge(HitsData3, AtBatsData2, on='Stadium')
```

```python
#Doing the calculation for batting average (Batting Average = Total Hits / Total At Bats)
BattingAverage['BA'] = BattingAverage['Total Hits']/BattingAverage['At Bats']
```

```python
BattingAverage.sort_values('BA',ascending = False)
```

```python
altitudes.columns = ['Stadium', 'Altitude']
BattingAverage = pd.merge(BattingAverage, altitudes, on='Stadium')
```

```python
BattingAverage
```

```python
BattingAverage.sort_values('BA',ascending = False)
```

```python
Corr = BattingAverage['BA'].corr(BattingAverage['Altitude'])
print("The correlation between BA and Altitude is ", Corr)
```

```python
x = BattingAverage['Altitude'].copy()
y = BattingAverage['BA'].copy()
fig, ax = plt.subplots()
scatter = ax.scatter(x,y,marker='o',label="Altitude vs BA")
ax.set_xlabel("Altitude", fontsize=16)
ax.set_ylabel("BA", fontsize=16)
ax.set_title("Altitude vs BattingAverage", fontsize = 16)
plt.legend()
```
