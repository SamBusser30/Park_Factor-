```python
# event_cd codes:
# 23 => Homerun
# 22 => Triple
# 21 => Double
# 20 => Single

events_with_altitudes = pd.merge(events, altitudes, on='home_team_id')
events_with_altitudes.head(10)
```

```python
HomeRunData = events_with_altitudes.drop(events_with_altitudes[events_with_altitudes.event_cd!=23].index)
HomeRunData['altitude'].hist(bins=150)
```

```python
HomeRunData['altitude'].plot()
```

```python
HomeRunData['altitude'].value_counts()
```

```python
homeruns_vs_stadium = HomeRunData['home_team_id'].value_counts()
homeruns_vs_stadium1 = homeruns_vs_stadium.plot(kind='bar', title="Stadium vs. Home Runs")
homeruns_vs_stadium1.set_ylabel("Number of Home Runs")
homeruns_vs_stadium1.set_xlabel("Stadium")
```

```python
homeruns_vs_altitude = HomeRunData['altitude'].value_counts()
homeruns_vs_altitude
homeruns_vs_altitude1 = homeruns_vs_altitude.sort_index(axis=0, ascending=True, kind='quicksort').plot(kind='bar', title="Home Runs vs. Altitude", color='g', legend=False)
homeruns_vs_altitude1.set_ylabel("Number of Home Runs")
homeruns_vs_altitude1.set_xlabel("Altitude")
```

```python
# event_runs_ct
#totalruns_vs_altitude = HomeRunData['altitude'].value_counts()
#totalruns_vs_altitude1 = totalruns_vs_altitude.sort_index(axis=0, ascending=True, kind='quicksort').plot(kind='bar', title="Home Runs vs. Altitude")
#totalruns_vs_altitude1.set_ylabel("Number of Home Runs")
#totalruns_vs_altitude1.set_xlabel("Altitude")

total_runs = [[0 for i in range(2)] for j in range(30)] 

for i in range(0,30):
    total_runs[i][0] = altitudes['altitude'][i]
    

for i in range(0,30):
    
    for j in range(0,2111526):
    
        if events_with_altitudes['altitude'][j] == total_runs[i][0]:
        
            total_runs[i][1] += events_with_altitudes['event_runs_ct'][j]
    
    
total_runs
    
```

```python
#import numpy as np 
#import matplotlib.pyplot as plt

#total_runs_array = np.array(total_runs)

#plt.imshow(total_runs_array)
#plt.show()

runs = pd.DataFrame(data=total_runs)
runs1 = runs.sort_values(0, ascending=True).plot.bar(x=0, y=1, title = 'Total Runs vs. Altitude', color='g', legend=False)
runs1.set_ylabel("Total Number of Runs")
runs1.set_xlabel("Altitude")
```

```python
scatter_homeruns_altitude = homeruns_vs_altitude.to_frame(name=None)
#scatter_homeruns_altitude
scatter_homeruns_altitude1 = scatter_homeruns_altitude.reset_index().plot(kind='scatter',x='index', y='altitude', title='Home Runs vs. Altitude')
scatter_homeruns_altitude1.set_xlabel("Altitude")
scatter_homeruns_altitude1.set_ylabel("Number of Home Runs")
#
```

```python
import seaborn as sns

#scatter_homeruns_altitude = scatter_homeruns_altitude.reset_index()

g1 = sns.regplot(scatter_homeruns_altitude['index'],scatter_homeruns_altitude['altitude'],scatter_kws={"color": "green"}, line_kws={"color": "red"})

# Set the Title of the graph from here
g1.axes.set_title('Home Runs vs. Altitude')
# Set the xlabel of the graph from here
g1.set_xlabel("Altitude")
# Set the ylabel of the graph from here
g1.set_ylabel("Number of Home Runs")
```

```python
# Single plots
SinglesData = events_with_altitudes.drop(events_with_altitudes[events_with_altitudes.event_cd!=20].index)

singles_vs_altitude = SinglesData['altitude'].value_counts()
singles_vs_altitude
singles_vs_altitude1 = singles_vs_altitude.sort_index(axis=0, ascending=True, kind='quicksort').plot(kind='bar', title="Singles vs. Altitude", color='b', legend=False)
singles_vs_altitude1.set_ylabel("Number of Singles")
singles_vs_altitude1.set_xlabel("Altitude")
```

```python
scatter_singles_altitude = singles_vs_altitude.to_frame(name=None)

scatter_singles_altitude1 = scatter_singles_altitude.reset_index().plot(kind='scatter',x='index', y='altitude', title='Singles vs. Altitude')
scatter_singles_altitude1.set_xlabel("Altitude")
scatter_singles_altitude1.set_ylabel("Number of Singles")
```
